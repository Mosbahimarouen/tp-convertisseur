Soit le bloc de code suivant :

```python
# Code de la fonction à tester : pour calculer la somme des diviseurs inférieurs à n
def somme_quoi_inferieurs(n):
    somme = 0
    for i in range(1,n/2,1):
        if n % i == 0:
            somme += i
    return somme
```

### Travail demandé :
Complétez les **tests unitaires** pour cette fonction.

- Tester un nombre premier  
- Tester un nombre parfait (ex: 6)  
- Tester un cas limite (n = 1)  
- Tester un type invalide (si pertinent)  

Utilisez `pytest` ou `unittest`.


````python
def somme_quoi_inferieurs(n):
    if not isinstance(n, int):
        raise TypeError("n doit être un entier")

    somme = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            somme += i
    return somme
    
````

````python
import pytest

def test_nombre_premier():
    assert somme_quoi_inferieurs(7) == 1

def test_nombre_parfait():
    assert somme_quoi_inferieurs(6) == 6

def test_cas_limite():
    assert somme_quoi_inferieurs(1) == 0

def test_type_invalide():
    with pytest.raises(TypeError):
        somme_quoi_inferieurs("abc")
````

````python
test_nombre_premier()
test_nombre_parfait()
test_cas_limite()
test_type_invalide()
print(" Tous les tests Exercice 1 passent")
````

## Exercice 2 (/5 pts)

Soit le bloc de code suivant :

```python
# Créez une fonction simple à tester utiliser la classe TestCase.
def addition(a,b):
    return a+b-1

# Test unitaire sans la classe
def test_addition():
    result = addition(2,3)
```

### Travail demandé :
Ajoutez les instructions de test unitaire au code en utilisant la classe `TestCase`.

- Créez une classe héritant de `unittest.TestCase`
- Écrivez au moins 3 tests
- Montrez qu’un test échoue à cause de l’erreur logique dans `addition`
- Proposez la correction de la fonction


````python
import unittest
# Fonction fournie (avec erreur logique)
def addition(a, b):
    return a + b - 1

class TestAddition(unittest.TestCase):

    def test_addition_simple(self):
        # Doit échouer (car 2+3 devrait donner 5)
        self.assertEqual(addition(2, 3), 5)

    def test_addition_zero(self):
        # Doit échouer (0+0 devrait donner 0)
        self.assertEqual(addition(0, 0), 0)

    def test_addition_negatif(self):
        # Doit échouer (-1+1 devrait donner 0)
        self.assertEqual(addition(-1, 1), 0)


suite = unittest.defaultTestLoader.loadTestsFromTestCase(TestAddition)
unittest.TextTestRunner(verbosity=2).run(suite)
    
````
